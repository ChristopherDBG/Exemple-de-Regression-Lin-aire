```python
#nouveau dataframe
# Liste des colonnes Ã  supprimer
cols_to_remove = ['Unnamed: 0','Category','Item','Serving Size','Total Fat','Total Fat (% Daily Value)','Saturated Fat (% Daily Value)','Sodium (% Daily Value)',
                 'Carbohydrates (% Daily Value)','Dietary Fiber (% Daily Value)']  # Remplacez par les noms des colonnes que vous souhaitez supprimer

# Supprimer les colonnes
menu1 = menus.drop(columns=cols_to_remove, axis=1)
```

```python
menus_1=menu1
menus_1= menus_1.drop(columns=['Vitamin C (% Daily Value)','Trans Fat'], axis=1)
menus_1.head()
```

```python
#matrice de corrÃ©lation
sns.heatmap(menus_1.corr(), center = 0, cmap = "PiYG")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import numpy as np
```

```python
len(menus_1.columns)
```

```python
#RÃ©aliser une ACP
wine_bis = scale(menus_1)
pca = PCA(n_components = 13)
pca.fit(wine_bis)
```

```python
#Choix du nombre de variables
eig = pandas.DataFrame(
    {
        "Dimension" : ["Dim" + str(x + 1) for x in range(13)], 
        "Valeur propre" : pca.explained_variance_,
        "% variance expliquÃ©e" : np.round(pca.explained_variance_ratio_ * 100),
        "% cum. var. expliquÃ©e" : np.round(np.cumsum(pca.explained_variance_ratio_) * 100)
    },
    columns = ["Dimension", "Valeur propre", "% variance expliquÃ©e", "% cum. var. expliquÃ©e"]
)
eig.head(4)
```

Commentaire: On veut que les deux premiÃ¨res variables permettent d'expliquer 72%  de nos donnÃ©es.

```python
#variance expliquÃ©e par dimension
plt.figure(figsize=(16, 6))
g_eig = sns.barplot(x = "Dimension", 
                        y = "% variance expliquÃ©e",
                        palette = ["lightseagreen"],
                        data = eig)
plt.text(12, 100/16, "Seuil")
plt.axhline(y = 100/13, linewidth = .5, color = "dimgray", linestyle = "--")
g_eig.set(ylabel = "Variance expliquÃ©e (%)")
g_eig.figure.suptitle("Variance expliquÃ©e par dimension")

plt.show()
```

```python
#crÃ©ation du nuages de point sur le plan facroriel
#ici on peut faire cette projection par la variable 'catÃ©gory'

wine_pca_df = pandas.DataFrame(pca.transform(wine_bis), 
                               columns = ["Dim" + str(i+1) for i in range(13)]) \
                    .assign(Class = menus["Category"])


fig, axes = plt.subplots(figsize = (16, 8))
fig.suptitle("Premier plan factoriel")
axes.axvline(x = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
axes.axhline(y = 0, color = 'lightgray', linestyle = '--', linewidth = 1)

g_pca = sns.scatterplot(x = "Dim1", y = "Dim2", hue = "Class", data = wine_pca_df, 
                            palette = "Set1", s = 100)
g_pca.set(xlabel = "Dimension 1", ylabel = "Dimension 2")

plt.show()
```

```python
#cercle de correlation
coordvar = pca.components_.T * np.sqrt(pca.explained_variance_)
coordvar_df = pandas.DataFrame(coordvar, columns=['PC'+str(i+1) for i in range(13)], index=menus_1.columns)
fig, axes = plt.subplots(figsize = (10, 10))
fig.suptitle("Cercle des corrÃ©lations")
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)
axes.axvline(x = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
axes.axhline(y = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
for j in range(13):
    axes.text(coordvar_df["PC1"][j],coordvar_df["PC2"][j], coordvar_df.index[j], size = 25)
    axes.plot([0,coordvar_df["PC1"][j]], [0,coordvar_df["PC2"][j]], color = "gray", linestyle = 'dashed')
plt.gca().add_artist(plt.Circle((0,0),1,color='blue',fill=False))

plt.show()
```
